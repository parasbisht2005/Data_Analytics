```python
import tkinter as tk
from tkinter import messagebox

# Menu and prices
menu = {
    'pizza': 60,
    'pasta': 40,
    'burger': 60,
    'salad': 70,
    'coffee': 80,
}

# Function to add item to the order
def add_to_order():
    item = item_var.get().lower()
    if item in menu:
        global order_total
        order_total += menu[item]
        order_list.append(item)
        order_label.config(text=f"Order: {', '.join(order_list)}")
        total_label.config(text=f"Total: {order_total} Rs")
    else:
        messagebox.showerror("Error", f"Sorry, we don't have {item} on the menu.")

# Function to finalize the order
def finalize_order():
    gst_rate = 0.12
    gst_amount = order_total * gst_rate
    final_total = order_total + gst_amount
    summary = (
        f"Order Summary:\n"
        f"Items: {', '.join(order_list)}\n"
        f"Order Total: {order_total} Rs\n"
        f"GST (12%): {gst_amount:.2f} Rs\n"
        f"Final Amount to Pay: {final_total:.2f} Rs"
    )
    messagebox.showinfo("Order Summary", summary)
    reset_order()

# Function to reset the order
def reset_order():
    global order_total, order_list
    order_total = 0
    order_list = []
    order_label.config(text="Order: None")
    total_label.config(text="Total: 0 Rs")

# Initialize GUI
root = tk.Tk()
root.title("Python Restaurant")

# Variables
order_total = 0
order_list = []
item_var = tk.StringVar()

# GUI Layout
tk.Label(root, text="Welcome to Python Restaurant", font=("Arial", 16)).pack(pady=10)
tk.Label(root, text="Menu", font=("Arial", 14)).pack()
menu_text = "\n".join([f"{item.capitalize()}: {price} Rs" for item, price in menu.items()])
tk.Label(root, text=menu_text, font=("Arial", 12)).pack(pady=10)

tk.Label(root, text="Enter the item you want to order:", font=("Arial", 12)).pack(pady=5)
tk.Entry(root, textvariable=item_var, font=("Arial", 12)).pack(pady=5)
tk.Button(root, text="Add to Order", command=add_to_order, font=("Arial", 12)).pack(pady=10)

order_label = tk.Label(root, text="Order: None", font=("Arial", 12))
order_label.pack(pady=5)

total_label = tk.Label(root, text="Total: 0 Rs", font=("Arial", 12))
total_label.pack(pady=5)

tk.Button(root, text="Finalize Order", command=finalize_order, font=("Arial", 12)).pack(pady=10)
tk.Button(root, text="Reset Order", command=reset_order, font=("Arial", 12)).pack(pady=5)

# Run the GUI
root.mainloop()
```
